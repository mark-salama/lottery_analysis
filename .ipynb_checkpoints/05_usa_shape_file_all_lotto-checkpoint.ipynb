```python
import geopandas as gpd
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from os.path import join
import os
from shapely.geometry import Point, Polygon
from uszipcode import SearchEngine
import imageio
```

```python
from PIL import Image, ImageOps
```

```python
# jupyter
cwd = os.getcwd()
```

```python
infile = join("cb_2017_us_state_500k")
```

```python
states_map = gpd.read_file(infile)
```

```python
states_map = states_map[states_map['NAME'] != 'Alaska']
```

```python
# all lottery
# load sales per capita from state level analysis
```

```python
all_lottery = pd.read_csv('total_lottery_sales_capita.csv')
```

```python
states_map_pop = states_map.merge(all_lottery,left_on='STUSPS',right_on='state')
```

```python
# add 0 for all states not in loaded csv
missing_states = states_map[~states_map.STUSPS.isin(states_map_pop.state.unique())]
missing_states['sales_per_capita_ex_video'] = 0
states_map_pop = pd.concat([states_map_pop,missing_states],axis=0)
```

```python
# cap sles per capita at 500 to add granularity becasue MA is 700+ and 
states_map_pop['adj_sales_per_capita_ex_video'] = states_map_pop.sales_per_capita_ex_video.apply(lambda x: min(x,500))
```

```python
# just to collect legend

f, ax = plt.subplots(1, figsize=(5,10))

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Helvetica']

ax.set_xlim([-127,-65])
ax.set_ylim([23,51])

states_map_pop.plot(ax=ax, linewidth=1, edgecolor='2', column='adj_sales_per_capita_ex_video',  cmap='viridis', legend=True)
```

```python
f, ax = plt.subplots(1, figsize=(40,40))

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Helvetica']

ax.set_xlim([-127,-65])
ax.set_ylim([23,51])
ax.set_axis_off()

ax.annotate('@marksalama', xy=(1500,150), xycoords='axes points',
        size=20, ha='right', va='center')

# for legend

ax.annotate('$500+', xy=(2355,815), xycoords='axes points',
        size=28, ha='right', va='center')

ax.annotate('$400', xy=(2331,672), xycoords='axes points',
        size=28, ha='right', va='center')

ax.annotate('$300', xy=(2331,529), xycoords='axes points',
        size=28, ha='right', va='center')

ax.annotate('$200', xy=(2331,386), xycoords='axes points',
        size=28, ha='right', va='center')

ax.annotate('$100', xy=(2331,243), xycoords='axes points',
        size=28, ha='right', va='center')

ax.annotate('$0', xy=(2295,100), xycoords='axes points',
        size=28, ha='right', va='center')

ax.annotate('source: state lotto financials, wikipedia', xy=(2400,10), xycoords='axes points',
        size=15, ha='right', va='center')

img = states_map_pop.plot(ax=ax, linewidth=1, edgecolor='2', column='adj_sales_per_capita_ex_video',  cmap='viridis')

img.get_figure().savefig('map.png')

img = Image.open("map.png")

legend = Image.open("legend.png") 

# paste legend
img.paste(legend, (2580, 1100))

w, h = img.size
img = img.crop((380, 700, w-100, h-920))

# position title in the center
w, h = img.size
title = Image.open("title.png") 
w_title, h_title = title.size
img.paste(title, (int(w/2-w_title/2), 50))

img.save("all_lotto.png",format='PNG')
```
